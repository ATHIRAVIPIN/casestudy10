```python
#encode the data set
from sklearn import preprocessing
lc=preprocessing.LabelEncoder()
mush_col=list(mush.columns)

for i in range(len(mush_col)):
    mush[mush_col[i]] = lc.fit_transform(mush[mush_col[i]])
```

```python
 # setting the target column
x=mush.drop(['class'],axis=1)
y=mush['class']
```

```python
#SPLITTING DATA 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=42)
```

```python
#MINMAX SCALING
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler(feature_range=(0,1))
x_train=mm.fit_transform(x_train)
x_test=mm.fit_transform(x_test)
```

```python
#LOGISTIC REGRESSION

from sklearn.linear_model import LogisticRegression
logit_model=LogisticRegression()
logit_model.fit(x_train,y_train)
y_pred_lm=logit_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score
print("Accuracy is ",accuracy_score(y_test,y_pred_lm))
print("f1 score is ",f1_score(y_test,y_pred_lm,average='macro'))
print("\nConfusion Matrix \n\n",confusion_matrix(y_test,y_pred_lm))
```

```python
#Decision Tree Algorithm

from sklearn.tree import DecisionTreeClassifier
dt_model=DecisionTreeClassifier()
dt_model.fit(x_train,y_train)
y_pred_dt=dt_model.predict(x_test)
```

```python
print("Accuracy is ",accuracy_score(y_test,y_pred_dt))
print("f1 score is ",f1_score(y_test,y_pred_dt,average='macro'))
print("\nConfusion Matrix \n\n",confusion_matrix(y_test,y_pred_dt))
```

```python
#SVM Model

from sklearn.svm import SVC
svm_linear=SVC(kernel='linear')
svm_linear.fit(x_train,y_train)
y_pred_svm=svm_linear.predict(x_test)
```

```python
print("Accuracy is ",accuracy_score(y_test,y_pred_svm))
print("f1 score is ",f1_score(y_test,y_pred_svm,average='macro'))
print("\nConfusion Matrix \n\n",confusion_matrix(y_test,y_pred_svm))
```

```python
#Random Forest Model

from sklearn.ensemble import RandomForestClassifier
r_forest=RandomForestClassifier()
r_forest.fit(x_train,y_train)
y_pred_rt=r_forest.predict(x_test)
```

```python
print("Accuracy is ",accuracy_score(y_test,y_pred_rt))
print("f1 score is ",f1_score(y_test,y_pred_rt,average='macro'))
print("\nConfusion Matrix \n\n",confusion_matrix(y_test,y_pred_rt))
```

```python
#KNN Model

from sklearn.neighbors import KNeighborsClassifier
acc_values=[]
neighbours=np.arange(3,15)
for k in neighbours:
    classifier=KNeighborsClassifier(n_neighbors=k,metric='minkowski')
    classifier.fit(x_train,y_train)
    y_pred=classifier.predict(x_test)
    acc=accuracy_score(y_test,y_pred)
    acc_values.append(acc)
acc_values
```

```python
plt.plot(neighbours,acc_values)
```

```python
classifier=KNeighborsClassifier(n_neighbors=5,metric='minkowski')
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)
```

```python
print("Accuracy is ",accuracy_score(y_test,y_pred))
print("f1 score is ",f1_score(y_test,y_pred,average='macro'))
print("\nConfusion Matrix \n\n",confusion_matrix(y_test,y_pred))
```

```python
# Naive Bayes Classifier - Guassian method

from sklearn.naive_bayes import GaussianNB
cl=GaussianNB()
cl.fit(x_train,y_train)
y_pred_g=cl.predict(x_test)
```

```python
print("ACCURACY SCORE =",accuracy_score(y_test,y_pred_g))
print("\nCONFUSION MATRIX =\n\n",confusion_matrix(y_test,y_pred_g))
```

```python
#Naive Bayes Classifier - Bernoulli method

from sklearn.naive_bayes import BernoulliNB
cl=BernoulliNB()
cl.fit(x_train,y_train)
y_pred_b=cl.predict(x_test)
```

```python
print("ACCURACY SCORE =",accuracy_score(y_test,y_pred_b))
print("\nCONFUSION MATRIX =\n\n",confusion_matrix(y_test,y_pred_b))
```

```python
#Naive Bayes Classifier - Multinominal method

from sklearn.naive_bayes import MultinomialNB
cl=MultinomialNB().fit(x_train,y_train)
y_pred_m=cl.predict(x_test)
```

```python
print("ACCURACY SCORE =",accuracy_score(y_test,y_pred_m))
print("\nCONFUSION MATRIX =\n\n",confusion_matrix(y_test,y_pred_m))
```
